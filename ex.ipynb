```python
class StrategyDeal:
    def __init__(self, bank, entry, targets, close):
        self.bank = bank
        self.entry = entry
        self.targets = targets
        self.close = close
        pass
        
    def get_targets(self):
        return self.targets


    def get_target_percents(self):
        return [round(100 * x / self.bank - 100, 3) for x in self.targets]
    

    def get_target_banks(self):
        return [bank * percent for percent in self.get_target_percents()]
        # список значений банков, если продавать активы по таргетам, как в пример, округленные до 3 знака [1069.12, 1133.764, 1168.573]

        
    def __str__(self):
        s = "BANK: {}".format(self.bank) + "\n"
        s += "START_PRICE: {}".format(self.entry) + "\n"
        s += "STOP_PRICE: {}".format(self.close) + "\n"
        targets = self.targets
        percents = self.get_target_percents()
        target_banks = self.get_target_banks()
        for i in len(targets):
            s += "\n"
            s += "{} target: {}".format(i, targets[i]) + "\n"
            s += "Percent: {}%".format(percents[i]) + "\n"
            s += "Bank: {}".format(target_banks[i]) + "\n"
        return s


def read_data(file_name):
    file = open(file_name)
    content = file.read()
    file.close()
    return content


def write_data(file_name, data):
    file = open(file_name, 'w')
    file.write(data)
    file.close() 
    
    
def parse_data(data, split_symbol="-----"):
    data_splitted = data.split(split_symbol)
    deals = []
    for entry in data_splitted:
        rows = entry.split("\n")
        bank_price = None
        entry_price = None
        close_price = None
        targets = []
        for row in rows:
            if row.startswith("Bank"):
                value_usd = row.split()[0]
                if value_usd.endswith("USD"):
                    bank_price = float(value_usd[0:-3])
            if row.startswith("Entry"):
                value_usd = row.split()[0]
                if value_usd.endswith("USD"):
                    entry_price = float(value_usd[0:-3])
            if row.startswith("Close"):
                value_usd = row.split()[0]
                if value_usd.endswith("USD"):
                    close_price = float(value_usd[0:-3]) 
            if row.startswith("Target"):
                row_splitted = row.split()
                for target_raw in row_splitted[1:]
                    if target_raw.endswith("USD;"):
                        targets.append(float(target_raw[0:-4]))
        if bank_price and entry_price and close_price and targets:
            deals.append({"bank": bank_price, "entry": entry_price, "close": close_price, "targets" = targets})
    return deals


def main():
    content = read_data('deals.txt')
    deals = parse_data(deals)
    result = ""
    for deal in deals:
        strat = StrategyDeal(deal["bank"], deal["entry"], deal["targets"], deal["close"])
        if result:
            result += "-----"
        result += str(strat)
    write_data('out.txt', result)


if __name__ == '__main__':    
    main() 
```
